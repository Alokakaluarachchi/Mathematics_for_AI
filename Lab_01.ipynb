```python
import numpy as np
```

1. Define two 2D vectors u and v:

```python
u = np.array([3, 2])
v = np.array([1, 4])
```

2.Compute vector operations

```python
print('u + v =', u + v)
print('u - v =', u - v)
print('2 * u =', 2 * u)
print('dot(u, v) =', np.dot(u, v))
```

3. Visualize the vectors using matplotlib

```python
import matplotlib.pyplot as plt
origin = [0], [0]
plt.quiver(*origin, u[0], u[1], color='r', label='u')
plt.quiver(*origin, v[0], v[1], color='b', label='v')
plt.quiver(*origin, (u+v)[0], (u+v)[1], color='g', label='u + v')
plt.quiver(*origin, (2*u)[0], (2*u)[1], color='m', label='2u')
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal')
plt.legend()
plt.title('2D Vector Operations')
plt.show()
```

Question 3: Unit Vectors and Cosine Similarity

1. Define two vectors x and y using NumPy

```python
import numpy as np
x = np.array([4, 3])
y = np.array([1, 2])
```

2. Calculate the magnitudes (norms) of x and y

```python
norm_x = np.linalg.norm(x)
norm_y = np.linalg.norm(y)
```

3.Find the unit vectors by dividing each vector by its magnitude

```python
unit_x = x / norm_x
unit_y = y / norm_y
```

4. Compute the cosine similarity using the dot product:

```python
cosine_similarity = np.dot(x, y) / (norm_x * norm_y)
print(f"Cosine Similarity: {cosine_similarity:.4f}")
```

value close to 1 means the vectors point in a similar direction
